```python
df = pd.read_csv("Chelsea_manager_ppg.csv")
```

```python
team = "Chelsea"

#iller = "lightsteelblue"
#rimary = "orangered"
background = "aliceblue"
text_color = "#01153E"






#getting data
Potter = df.Potter.tolist()
Tuchel = df.Tuchel.tolist()
Lampard = df.Lampard.tolist()
Sarri = df.Sarri.tolist()
Conte = df.Conte.tolist()
Hiddink = df.Hiddink.tolist()
Mourinho = df.Mourinho.tolist()
Benitez = df.Benitez.tolist()
DiMatteo = df.Di_Matteo.tolist()
VillasBoas = df.Villas_Boas.tolist()
Ancelotti = df.Ancelotti.tolist()

Games = df.Games.tolist()

Potter = [x for x in Potter if str(x) != 'nan']
Tuchel = [x for x in Tuchel if str(x) != 'nan']
Lampard = [x for x in Lampard if str(x) != 'nan']
Sarri = [x for x in Sarri if str(x) != 'nan']
Conte = [x for x in Conte if str(x) != 'nan']
Hiddink = [x for x in Hiddink if str(x) != 'nan']
Mourinho = [x for x in Mourinho if str(x) != 'nan']
Benitez = [x for x in Benitez if str(x) != 'nan']
DiMatteo = [x for x in DiMatteo if str(x) != 'nan']
VillasBoas = [x for x in VillasBoas if str(x) != 'nan']
Ancelotti = [x for x in Ancelotti if str(x) != 'nan']

def accumu(lis):
    total = 0
    for x in lis:
        total += x
        yield total

Potter = list(accumu(Potter))
Tuchel = list(accumu(Tuchel))
Lampard = list(accumu(Lampard))
Sarri = list(accumu(Sarri))
Conte = list(accumu(Conte))
Hiddink = list(accumu(Hiddink))
Mourinho = list(accumu(Mourinho))
Benitez = list(accumu(Benitez))
DiMatteo = list(accumu(DiMatteo))
VillasBoas = list(accumu(VillasBoas))
Ancelotti = list(accumu(Ancelotti))


#creating figure and axes
fig,ax = plt.subplots(figsize=(8,15))

#adding a grid
ax.grid(ls="dotted",lw=0.5,zorder=1)

ax.plot(Games[:17],Potter)
ax.plot(Games[:63],Tuchel,color = '#008080',label="Tuchel")
ax.plot(Games[:57],Lampard,color = '#054907',label="Lampard")
ax.plot(Games[:38],Sarri)
ax.plot(Games[:76],Conte,color = '#800000',label="Conte")
ax.plot(Games[:21],Hiddink)
ax.plot(Games,Mourinho,color = '#01153E',label="Mourinho")
ax.plot(Games[:26],Benitez)
ax.plot(Games[:23],DiMatteo)
ax.plot(Games[:27],VillasBoas)
ax.plot(Games[:76],Ancelotti, color = '#DAA520',label="Ancelotti")

#Legend
leg = plt.legend(loc='upper left')
for text in leg.get_texts():
    text.set_size(14)

s= "Chelsea PL Manager Points Won in Premier League\n"
htext.fig_text(0.14,1.05,s,color = text_color,fontproperties=title_font3, size = 16,style = "oblique")
#ig.text(0.14,1,"English Premier League".format(team),
 #       color=text_color,fontproperties=title_font,size = 14)

    
#adding x and y labels
ax.set_xlabel("Games Played",fontweight="bold",fontproperties=title_font3,size=14)
ax.set_ylabel("Points Won",fontproperties=title_font3, size=14,weight="bold")

#remove pips
ax.tick_params(axis="both",length=0, colors = text_color)

#set colour of spines
ax.spines['bottom'].set_color(text_color)
ax.spines['left'].set_color(text_color)

#adding logo
ax2= fig.add_axes([0.0,0.96,0.15,0.15])
ax2.axis("off")
url = "https://logos-world.net/wp-content/uploads/2020/05/Chelsea-Logo.png"
response = requests.get(url)
img = Image.open(BytesIO(response.content))
ax2.imshow(img)

#adding credits
fig.text(0.53,-0.025,"Created by Ewan Barclay. Data provided by football-data.co.uk",
        fontstyle="italic",fontproperties=italic_font3,size=8,color=text_color)

plt.tight_layout()
plt.savefig("plots/Chelsea_gd_RobotoMono", bbox_inches="tight", facecolor=background, edgecolor=background, dpi=300)
plt.show()
```
